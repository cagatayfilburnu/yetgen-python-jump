# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda():


    def __init__(self, tv_situation="Closed", tv_sound=0, channel_list=["FOX"], channel="FOX"):
        self.tv_situation = tv_situation
        self.tv_sound = tv_sound
        self.channel_list = channel_list
        self.channel = channel

    def tv_open(self):
        if (self.tv_situation == "Opened"):
            print("TV is already openned!")

        else:
            print("TV is openning...")
            self.tv_situation  = "Openned"

    def tv_close(self):
        if (self.tv_situation == "Closed"):
            print("TV is already closed")

        else:
            print("TV is closing...")
            self.tv_situation = "Closed"

    def volume_settings(self):
        while True:
            response =  input("Volume down: '+'\nVolume up: '-'\nQuit: quit")
            if (response == "-"):
                if (self.tv_sound != 0):
                    self.tv_sound -= 1
                    print("Ses:", self.tv_sound)

            elif (response == "+"):
                if (self.tv_sound != 31):
                    self.tv_sound += 1

                    print("Sound:", self.tv_sound)
            else:
                print("Volume is updated:", self.tv_sound)
                break

    def channel_add(self, channel_list):
        print("Channel is adding to TV list...")
        time.sleep(2)
        self.channel_list.append(channel_list)
        print("Channel is added.")

    def random_channel(self):
        random = random.randint(0, len(self.channel_list)-1)
        self.channel = self.channel_list[random]
        print("Channel is:" ,self.channel)

    def __len__(self):
        return len(self.channel_list)

    def __str__(self):
        return "TV Situation: {}\nTV Sound: {}\nChannel List: {}\nChannel: {}\n".format(self.tv_situation, self.tv_sound, self.channel_list, self.channel)


kumanda = Kumanda()


print("""

Welcome back to your best TV!


1. TV Open

2. TV Close

3. Volume Settings

4. Channel add

5. All channels

6. Random channel

7. TV Informations

Please  enter 'q' if you want to quit.
""")


while True:
    operation = input("Please enter a operation: ")

    if (operation == "q"):
        print("Leaving...")
        break

    elif (operation == "1"):
        kumanda.tv_open()

    elif (operation == "2"):
        kumanda.tv_close()

    elif (operation == "3"):
        kumanda.volume_settings()

    elif (operation == "4"):
        channels_name = input("Please add channels with ',':")

        channel_list = channels_name.split(",")

        for new in channel_list:
            kumanda.channel_add(new)

    elif (operation == "5"):
        print("Total Channel:",len(kumanda))

    elif (operation == "6"):
        kumanda.random_channel()

    elif (operation == "7"):
        print(kumanda)

    else:
        print("This is a wrong operation. Please enter a correct one!")
```
