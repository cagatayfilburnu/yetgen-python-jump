# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız
def word_changer(old_word, new_world):
    with open("products.txt", "r+") as file:
        content = file.read()
         # Belirtilen eski kelimeyi, yeni kelime ile değiştir
        content = content.replace(old_word, new_world)
        # Dosya imlecinin başına git
        file.seek(0)
        # Dosyayı temizle
        file.truncate()
        # Değiştirilmiş içeriği dosyaya yaz
        file.write(content)
        print("Changing process is completed")

word_changer("elma", "Armut")
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, model, speed, mileage):
        # Attributes
        self.model = model
        self.speed = speed
        self.mileage = mileage

class Bus(Vehicle):
    pass


vehicle_one = Bus("School Volvo", 180, 12)
print(f"Name: {vehicle_one.model}, Speed: {vehicle_one.speed}, Mileage: {vehicle_one.mileage}")
````
